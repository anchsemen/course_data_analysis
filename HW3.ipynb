Самое распространненые мужское и женское имени на корабле

```python
def extract_name(name):
    if '(' in name:
        return name.split('(')[1].split(' ')[0]
    else:
        return name.split(',')[1].split('.')[1].split(' ')[1]

df['First_Name'] = df['Name'].apply(extract_name)

male_passengers = df[df['Sex'] == 'male']
female_passengers = df[df['Sex'] == 'female']

popular_male_name = male_passengers['First_Name'].value_counts().idxmax()
popular_female_name = female_passengers['First_Name'].value_counts().idxmax()

print("Самое популярное мужское имя на корабле:", popular_male_name)
print("Самое популярное женское имя на корабле:", popular_female_name)
```

```python
def most_popular_name(group):
    return group['First_Name'].value_counts().idxmax()
```

```python
grouped_data = df.groupby(['Pclass', 'Sex'])

most_popular_names = grouped_data.apply(most_popular_name)

print("Самое популярное мужское и женское имя в каждом классе:")
print(most_popular_names)
```
